```python
# Importing necessary packages
import pandas as pd
import json
```

# Data Pre-processing

```python
# Reading supplier data from JSON file and loading it line-by-line as it is in the form of 'list of dictionaries'
with open('supplier_car.json', 'r') as handle:
    json_data = [json.loads(line) for line in handle]    
```

```python
# Creating Pandas DataFrame to try different data operations on it, and displaying the data sample with .head() method
data_df = pd.DataFrame(json_data)
data_df.head()
```

# Data Analysis

```python
# To see how the data is inside the DataFrame, applying 'groupby' technique on ID (column) with
# 'unique ID', and counting the number of entries in each group i.e. about 19 in total, for each group
data_df_gb = data_df.groupby('ID')
data_df_gb.count()
```

```python
# Getting a sample group to visualize particular attributes in the data
data_df_gb.get_group('4.0')
```

# Data Integration

```python
# To integrate all the attributes with each 'object ID', started by getting all these attributes in the form of lists i.e.
# 19 headings, added these lists with the particular attribute-name to the DataFrame and then dropped the duplicat headings
# with the same ID
a = 0
att_dict = {}
output_data = data_df.copy()

# Applying a loop on all the groups and taking two columns of interest from it i.e. 'Att-Names' and 'Att-Values'
for group in data_df_gb:
    dt = group[1][['Attribute Names','Attribute Values']]
    
    # On the first group, taking names of attributes and saving them in a list
    # Initializing a dictionary of attributes to store attributes in each iteration for particular 'Att-names'
    if a == 0:
        a+=1
        att_names = list(group[1]['Attribute Names'])
        for name in att_names:
            att_dict[name] = []
    
    # In each group, I tried to find out all the attribute names, took particular data to the specific and relevant attribute,
    # and stored in the attributes dictionary. If any particular attribute was not found in the group, I added 'null' to that
    # dictionary
    for att in att_names:
        try:
            att_dict[att].append(dt.loc[dt['Attribute Names'] == att, 'Attribute Values'].iloc[0])
        except:
            att_dict[att].append('')

# Now at this stage, I have all the attributes for all the object-id's in a dictonarry, and I am ready to add them to DataFrame,
# so I get can integrated data for each object_id in the dataset
```

```python
# Save the output data as a copy of the input data for now
output_data = data_df.copy()

# Delete the columns: Att-Names and Att-Values, as we are going to add them as a column for each attribute
del(output_data['Attribute Names'])
del(output_data['Attribute Values'])

# Drop duplicate data with ID, as the changing data in each duplicate entry were attributes while we are adding them as seperate
# columns for each ID
output_data = output_data.drop_duplicates(subset='ID')

# Add all the attributes columns to the ouput dataset
for name in att_names:
    output_data[name] = att_dict[name]

# Show the data-head to see the sample
output_data.head()
```

# Data Normalization

```python
# Data Normalization:
# Normalization is done on two columns as an example. The target columns are 'TransmissionTypeText' and 'FuelTypeText'.
# For 'TransmissionTypeText', I have normalized automatic to 1 and manual to 0 respectively; for 'FuelTypeText', I have
# normalized Benzin to 1 and Diesel to 0 respectively

norm_data = output_data.copy()
norm_data.loc[norm_data['TransmissionTypeText'].isin(['Automat','Automatisiertes Schaltgetriebe','Automatik-Getriebe','Automat sequentiell']) , 'TransmissionTypeText'] = 1
norm_data.loc[norm_data['TransmissionTypeText'].isin(['Schaltgetriebe','Schaltgetriebe manuell']) , 'TransmissionTypeText'] = 0

norm_data.loc[norm_data['FuelTypeText'].isin(['Benzin']) , 'FuelTypeText'] = 1
norm_data.loc[norm_data['FuelTypeText'].isin(['Diesel']) , 'FuelTypeText'] = 0

# Showing the data sample of normalized data
norm_data.head()
```

# Generating & Saving Data in Excel File Format

```python
# Now I will create an Excel file (worksheet) with three tabs: pre-processed data, integrated data and normalized data
# Initializing the Excel writer with Pandas to store data in it
writer = pd.ExcelWriter('Target Data.xlsx', engine='xlsxwriter')

# Write each DataFrame to a different tab in the Excel worksheet
data_df.to_excel(writer, sheet_name='Pre-processing')
output_data.to_excel(writer, sheet_name='Integration')
norm_data.to_excel(writer, sheet_name='Normalisation')

# Save the Pandas Excel writer to get the desired file i.e. 'Target Data.xlsx'
writer.save()
```
